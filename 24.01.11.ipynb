```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)

axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('그래프')
plt.show()
```

```python
#---------------------------------mission2-----------------------------------------
import matplotlib.pyplot as plt
import pandas as pd

plt.rc('font', family='NanumGothic')
df=pd.read_excel("data1.xlsx")
df.columns

plt.plot(df['kor'], 'g--', marker='o', label='국어점수')
plt.plot(df['math'],'r-', marker='^', label='수학점수')
plt.xticks(range(len(df.index)), labels=df['name'])
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적 그래프')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv('iris.csv')
print(iris.info(), end='\n')

print(iris.index)

print(iris.columns)
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title('Scatter Plpt by seaborn', fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='species',data=iris)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x='species',y='petal_length',data=iris, errorbar='sd')
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
sns.regplot(x='petal_length', y='petal_width', data=iris)
plt.show()
```

```python
df1=iris.corr()
sns.heatmap(data=df1,annot=True,fmt='.2f', cbar=False)
plt.show()
```
